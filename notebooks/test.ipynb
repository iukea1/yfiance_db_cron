```python
import yfinance as yf
dat_ = yf.Ticker("MSFT")
# Get historical data into a DataFrame
df = dat_.history(period="max")  # Gets 1 month of data by default
print(df.head())  # Display first few rows


# Get splits and dividends data
splits_df = dat_.splits
dividends_df = dat_.dividends
analyst_price_targets_df = dat_.analyst_price_targets


# Print the results
print("Stock Splits:")
print(splits_df.tail())
print("\nDividends:")
print(dividends_df.tail())

print(analyst_price_targets_df)
```
